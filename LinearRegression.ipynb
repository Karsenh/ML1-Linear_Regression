```python
# import pandas to get and describe table data
import pandas

# import DataFrame to parse and manipulate table
from pandas import DataFrame

# import matplotlib to create scatter plot
import matplotlib.pyplot as plt
```

```python
# Use 'pandas' to read into data csv file
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
# Use describe method to analyze data retrieved by pandas from csv file
data.describe()
```

```python
# Use DataFrame to parse data and separate columns by heading for x & y axis
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns =['worldwide_gross_usd'])
```

```python
X, y
```

```python
# Use matplotlib.pyplot (as plt) to create a scatter plot using the X and y axis from DataFrame
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs. Global Revenue $')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)

plt.show()
```
